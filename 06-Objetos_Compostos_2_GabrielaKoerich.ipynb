## 1 – Um objeto composto multidimensional (matriz) pode receber  objeto composto unidimensional (vetor)?

Sim, um objeto composto multidimensional, como uma matriz, pode receber um objeto composto unidimensional, como um vetor, desde que haja compatibilidade entre suas dimensões. Isso ocorre porque o vetor pode ser inserido em uma das dimensões da matriz, como em uma linha ou coluna, desde que o número de elementos seja correspondente. Caso as dimensões não sejam compatíveis, a atribuição não será possível.
