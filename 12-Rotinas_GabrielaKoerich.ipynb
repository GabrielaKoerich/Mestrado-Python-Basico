## Responda a pergunta do exercício 2 da aula 10, usando funções:<br>
Instancie um objeto da classe NumPy array, contendo 5 valores quantitativos randômicos normalmente distribuídos, com média 10 e desvio padrão 3.<br><br>
Quantos e quais são os valores superiores a 10?

## 02 – Descreva o racional da sua resposta:

R.: O programa deve começar com a instanciação de um array com valores aleatórios a partir de uma distribuição normal, com média 10 e desvio padrão 3. Depois disso, é definida uma função que recebe um array como entrada. O primeiro passo dentro da função é identificar e trabalhar com a estrutura do objeto recebido, assumindo que ele é um objeto composto do tipo numpy.ndarray. A função então aplica uma condição lógica diretamente sobre o array, selecionando apenas os elementos que são maiores que 10. Esse processo gera um novo array contendo somente os valores que atendem ao critério estabelecido. Na sequência, o programa calcula a quantidade de elementos desse novo array filtrado utilizando o atributo .size, que retorna o número de elementos presentes. Por fim, a função imprime a quantidade de valores maiores que 10 e quais são esses valores.

## 04 – Escreva o programa em Python:

```python
import numpy as np

exercicio2 = np.random.normal(loc=10, scale=3, size=5)
print(exercicio2)
```

```python
def imprimir_maiores_que_10(array_p_analise: np.ndarray) -> None:
    maiores = array_p_analise[array_p_analise > 10]
    print(f"Quantidade de elementos maiores que 10: {maiores.size}")
    print(f"Elementos maiores que 10: {maiores}")
```

```python
imprimir_maiores_que_10(array_p_analise=exercicio2)
```

## Desenvolva um programa de computador que retorne a média de um conjunto de valores.

## 06 – Descreva o racional da sua resposta:

R.: Para começar será criada uma função chamada calcular_media, que recebe como entrada um array. Dentro da função, o programa usa o método .mean() do numpy para calcular a média dos valores. Depois de obter o valor da média, ele imprime esse valor na tela com uma mensagem explicativa. A função não retorna nenhum valor, pois o objetivo principal é apenas mostrar o resultado, e não utilizá-lo em outro cálculo.

## 07 – Desenhe o fluxograma da resposta:

## 07 – Escreva o programa em Python:

```python
def calcular_media(array_p_analise: np.ndarray) -> None:
    media = array_p_analise.mean()
    print(f"Média dos elementos do array: {float(media)}")
    return None
```

```python
exemplo_array = np.array([1, 2, 3, 4, 5])
calcular_media(exemplo_array)
```

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def dividir_objetos_simples (divisor: int, dividendo: int) -> None:
    resultado = divisor / dividendo
    print(f"O divisor é do tipo: {type(divisor)}")
    print(f"O dividendo é do tipo: {type(dividendo)}")
    print(f"Resultado da divisão: {resultado}")
    return None
```

```python
dividir_objetos_simples(divisor=10, dividendo=2)
```
