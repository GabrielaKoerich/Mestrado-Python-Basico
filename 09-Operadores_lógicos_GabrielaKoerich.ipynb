## 01 – Dados as operações abaixo, qual o resultado da operação lógica: 
p and q and r and s and t?
<br>
<br>Como os operadores lógicos podem ser usados para inverter o resultado?

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

O resultado da operação lógica p and q and r and s and t é falso, pois p e t são falsos, e no operador “and” basta um valor falso para que toda a expressão seja falsa. Para inverter esse resultado, é possível usar o operador lógico “not”, fazendo com que a expressão passe a ser verdadeira. Outra possibilidade é utilizar o operador "or", que faria com que o resultado fosse resultado, já que bastaria apenas um desses valores serem verdadeiros para que se retorne o valor True.

```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2
p & q & r & s & t
```

```python
not (p & q & r & s & t)
```

```python
p or q or r or s or t
```

## 02 – Usando apenas o operador or, qual seria o resultado final das operações abaixo? Como deve ser usado o not para inverter o resultado?

<br>p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
<br>q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
<br>r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
<br>s = ['a',1,True,'b',0,False] > ['a',True,1]
<br>t = ['a',1,True,'b',0,False] == 1
<br>u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

Ao avaliar cada uma das expressões, observa-se que p, q, r, s e u resultam em valores verdadeiros, enquanto apenas t é falso. Como o operador lógico or retorna verdadeiro quando pelo menos uma das condições é verdadeira, a expressão completa p or q or r or s or t or u terá como resultado final verdadeiro, já que há vários valores verdadeiros envolvidos. Para inverter esse resultado, utiliza-se o operador not, aplicando-o à expressão completa. Dessa forma, como o resultado original é verdadeiro, a expressão not (p or q or r or s or t or u) passa a ser falsa.

```python
p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
s = ['a',1,True,'b',0,False] > ['a',True,1]
t = ['a',1,True,'b',0,False] == 1
u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

p or q or r or s or t or u
```

```python
not (p or q or r or s or t or u)
```
