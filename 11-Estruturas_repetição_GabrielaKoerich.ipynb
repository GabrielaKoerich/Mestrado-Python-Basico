## Instancie um objeto da classe lista contendo 5.000 valores inteiros iniciando em 0. 
Quantos são pares e quantos são impares?

## 01 – Descreva o racional da sua resposta:

R.: O programa tem como objetivo contar quantos números pares e ímpares existem em uma sequência de 0 até 4999. Inicialmente, ele cria uma lista vazia para armazenar os valores e define duas variáveis contadoras, uma para números pares e outra para ímpares, ambas iniciadas em zero. Em seguida, utiliza um laço de repetição para percorrer todos os números dentro desse intervalo e, a cada iteração, adiciona o número atual à lista. Para cada valor, o programa verifica se ele é divisível por dois por meio da operação de módulo; caso o resto da divisão seja zero, o número é classificado como par e o contador de pares é incrementado, caso contrário, ele é considerado ímpar e o contador correspondente é incrementado. Ao final do processo, depois de analisar todos os números, o programa exibe na tela a quantidade total de números pares e ímpares encontrados, que nesse caso são iguais, já que o intervalo possui a mesma quantidade de valores pares e ímpares.

## 03 – Escreva o programa em Python:

```python
import pandas as pd
exercicio1aula11 = []
x=0
y=0
for i in range(0, 5000):
    exercicio1aula11.append(i)
    if exercicio1aula11[i] % 2 == 0:
        x += 1
    else:
        y += 1
print(f"Quantidade de números pares: {x}")
print(f"Quantidade de números ímpares: {y}")
```

## 05 – Descreva o racional da sua resposta:

R.: O programa tem como objetivo separar os números de um conjunto em pares e ímpares, além de contar quantos existem de cada tipo. Inicialmente, são criadas duas listas vazias, uma para armazenar os números pares e outra para os ímpares, juntamente com dois contadores iniciados em zero. Em seguida, o programa percorre todos os valores do conjunto de dados e, para cada número, verifica se ele é divisível por dois. Caso seja, ele é classificado como par, sendo adicionado à lista correspondente e incrementando o contador de pares; caso contrário, é classificado como ímpar, sendo armazenado na outra lista e incrementando o respectivo contador. Ao final da execução, o programa apresenta na tela a quantidade de números pares e ímpares, bem como as listas contendo os valores de cada grupo.

## 07 – Escreva o programa em Python:

```python
import pandas as pd
exercicio7aula11 = pd.read_csv('dataframe_genes.csv')
exercicio7aula11 = exercicio7aula11.set_index('Nome do gene')
exercicio7aula11.drop("Unnamed: 0", axis=1, inplace=True)

exercicio7aula11 
```

```python
listapares = []
listaimpares = []
b = 0
c = 0

for i in range(len(exercicio7aula11)):
    if exercicio7aula11.iloc[i, 0] % 2 == 0:
        listapares.append(int(exercicio7aula11.iloc[i, 0]))
        b += 1
    else:
        listaimpares.append(int(exercicio7aula11.iloc[i, 0]))
        c += 1

print(f"Quantidade de números pares: {b}")
print("Números pares:", listapares)
print(f"Quantidade de números ímpares: {c}")
print("Números ímpares:", listaimpares)
```

## 08 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
listapares = []
listaimpares = []
d = 0
e = 0

i = 0
while i < len(exercicio7aula11):
    if exercicio7aula11.iloc[i, 0] % 2 == 0:
        listapares.append(int(exercicio7aula11.iloc[i, 0]))
        d += 1
    else:
        listaimpares.append(int(exercicio7aula11.iloc[i, 0]))
        e += 1
    i += 1

print(f"Quantidade de números pares: {d}")
print("Números pares:", listapares)
print(f"Quantidade de números ímpares: {e}")
print("Números ímpares:", listaimpares)
```

## 09 – Usando a coluna de presença de hu no nome do gene, itere os index deste objeto usando a função enumerate(). Estes index podem ser usados para imprimir cada um dos valores presentes na coluna dos dois primeiros caracteres dos nomes dos genes?

Sim, os índices obtidos por meio da função enumerate() podem ser utilizados para imprimir os valores presentes na coluna correspondente aos dois primeiros caracteres dos nomes dos genes. Ao aplicar enumerate() sobre a coluna de presença de “HU no nome do gene”, o programa percorre cada elemento dessa coluna ao mesmo tempo em que fornece o índice associado a cada linha. Esse índice corresponde à posição da linha no DataFrame e pode ser utilizado para acessar diretamente o índice do DataFrame, que contém os nomes dos genes. Dessa forma, ao utilizar esse índice com um fatiamento, como nos dois primeiros caracteres, é possível recuperar e imprimir corretamente os valores desejados. Isso demonstra que os índices gerados por enumerate() são úteis não apenas para iterar sobre os valores de uma coluna, mas também para acessar e manipular outras informações relacionadas à estrutura do DataFrame.

```python
for i, _ in enumerate(exercicio7aula11['HU no nome do gene']):
    print(exercicio7aula11.index[i][:2])
```

## 10 – Selecione qualquer uma das instâncias contidas na DataFrame. É possível recuperar as chaves e os valores com ao menos 3 estruturas de repetição diferentes?

Sim, é possível recuperar as chaves e os valores de uma instância de um DataFrame utilizando pelo menos três estruturas de repetição diferentes. Para isso, inicialmente seleciona-se uma linha do DataFrame, por exemplo, utilizando o método loc, que retorna uma série contendo os valores daquela instância e os nomes das colunas como chaves. A partir dessa seleção, pode-se percorrer os dados utilizando um laço for com o método items(), que permite iterar diretamente sobre pares de chave e valor. Outra alternativa é utilizar um laço for com a função range, acessando os elementos por meio de índices, o que exige o uso das posições e dos rótulos das colunas. Além disso, também é possível empregar um laço while, controlando manualmente um índice para percorrer todos os elementos da instância. 

```python
linha = exercicio7aula11.loc["BRCA1"]

for chave, valor in linha.items():
    print(f"{chave}: {valor}")
```

```python
linha = exercicio7aula11.loc["BRCA1"]

for i in range(len(linha)):
    print(f"{linha.index[i]}: {linha[i]}")
```

```python
linha = exercicio7aula11.loc["BRCA1"]

i = 0
while i < len(linha):
    print(f"{linha.index[i]}: {linha[i]}")
    i += 1
```
