```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

if p and q and r and s and t:
    print("Todas as condições são verdadeiras.")
else:    print("Pelo menos uma condição é falsa.")
```

## 04 – Escreva o programa em Python:

```python
import numpy as np

exercicio2 = np.random.normal(loc=10, scale=3, size=5)
print(exercicio2)
```

```python
x=0
lista =[]
for i in range(len(exercicio2)):
            if exercicio2[i] > 10:
                lista.append(float(exercicio2[i]))
                x = x + 1
print(f"Elementos maiores que 10: {lista}")
print(f"Número de elementos maiores que 10: {x}") 
```

Em seguida, a coluna “Tamanho do gene” foi convertida para tipo numérico, permitindo a realização de comparações. A análise baseou-se no tamanho do primeiro gene do DataFrame. A partir desse valor, foi aplicada uma estrutura condicional (if/elif) com três cenários: (i) se o tamanho estiver entre 2 e 4, são exibidas as informações do próprio gene; (ii) se for menor que 2, são mostrados os genes com tamanho maior que 4; e (iii) se for maior que 4, são exibidos os genes com o maior tamanho presente no conjunto de dados. A seleção dos dados foi feita por meio de filtragem direta no DataFrame, preservando o formato tabular e garantindo eficiência na manipulação dos dados.

## 09 – Escreva o programa em Python:

```python
import pandas as pd
exercicio7 = pd.read_csv('dataframe_genes.csv')
exercicio7 = exercicio7.set_index('Unnamed: 0')


exercicio7 = exercicio7.T
exercicio7.columns = [
    'Nome do gene',
    'Tamanho do gene',
    'Início do nome do gene',
    "Número de A's no nome do gene",
    'HU no nome do gene'
]
exercicio7 
```

```python
exercicio7['Tamanho do gene'] = pd.to_numeric(exercicio7['Tamanho do gene'])
primeiro_gene = exercicio7.index[0]
tamanho_primeiro = exercicio7.loc[primeiro_gene, 'Tamanho do gene']
```

```python
print(f"Primeiro gene: {primeiro_gene}")
print(f"Tamanho do primeiro gene: {tamanho_primeiro}\n")

if 2 <= tamanho_primeiro <= 4:
    print("✔ O tamanho está entre 2 e 4 (inclusive).")
    print("→ Mostrando informações do primeiro gene:\n")
    print(exercicio7.loc[primeiro_gene])

elif tamanho_primeiro < 2:
    print("⚠ O tamanho do primeiro gene é menor que 2.")
    print("→ Mostrando genes com tamanho MAIOR que 4:\n")
    print(exercicio7[exercicio7['Tamanho do gene'] > 4])

elif tamanho_primeiro > 4:
    print("⚠ O tamanho do primeiro gene é maior que 4.")
    
    maior = exercicio7['Tamanho do gene'].max()
    
    print(f"→ O maior tamanho encontrado é: {maior}\n")
    print("→ Gene(s) com tamanho maior que 4:\n")
    resultado = exercicio7[exercicio7['Tamanho do gene'] == maior]
    display(resultado)
 
```

## 10 – Caso a DataFrame contenha 10 milhões de entradas, este racional utilizado poderia responder as mesmas perguntas? Porque?
