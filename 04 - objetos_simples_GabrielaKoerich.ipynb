# Exercício 01:

No paradigma de Programação Orientada a Objetos, o que é um Objeto?
<br>
<br>
R.: Na Programação Orientada a Objetos, um objeto é uma instância de uma classe que possui atributos e métodos, representando tanto características quanto comportamentos de uma entidade. A classe funciona como um modelo ou molde, enquanto o objeto é a concretização desse modelo na prática. Os atributos correspondem aos dados ou estado do objeto, e os métodos às ações que ele pode executar. Assim, o objeto reúne informações e funcionalidades em uma única estrutura.
