```python
ultima_linha = lista3.iloc[-1]
ultima_coluna = lista3.iloc[:, -1]

ultima_linha_t = lista3.iloc[[-1]].transpose()

print("Última linha:")
print(ultima_linha.describe())

print("\nÚltima coluna:")
print(ultima_coluna.describe())
```

Embora a última linha e a última coluna apresentem a mesma média (5.015), a última coluna possui maior variabilidade, evidenciada pelo maior desvio padrão e pela maior amplitude entre os valores mínimo e máximo.
